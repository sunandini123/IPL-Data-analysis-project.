```python
## Common Positions with Examples:


loc Value	bbox_to_anchor=(x, y)	Meaning
'upper left'	(0, 1)	Top-left inside plot
'upper right'	(1, 1)	Top-right inside plot
'lower left'	(0, 0)	Bottom-left inside plot
'lower right'	(1, 0)	Bottom-right inside plot
'center'	(0.5, 0.5)	Center of the plot
Any loc value	(1.05, 1)	Just outside the top-right corner
```
